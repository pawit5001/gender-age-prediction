```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt  # สำหรับแสดงรูปภาพ

# โหลดโมเดลที่ฝึกเสร็จแล้ว
model = tf.keras.models.load_model('smart_classification.keras')

# ฟังก์ชันสำหรับเตรียมข้อมูลภาพ
def prepare_image(img_path):
    img = image.load_img(img_path, target_size=(224, 224))  # ขนาดที่โมเดลต้องการ
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)  # เพิ่มมิติ batch
    img_array /= 255.0  # ปรับค่ารูปภาพให้เป็น 0-1
    return img_array
```

```python
# ฟังก์ชันทำนายเพศ
def predict_gender(img_path):
    img_array = prepare_image(img_path)

    # ทำนายผล
    prediction = model.predict(img_array)
    prediction_value = prediction[0][0]  # ดึงค่าที่ทำนายได้

    # แสดงรูปภาพ
    img = image.load_img(img_path)
    plt.imshow(img)
    plt.axis('off')  # ปิดแกน
    plt.show()

    # แสดงค่า prediction ที่ได้เป็นตัวเลข
    print(f'Prediction Value: {prediction_value:.4f}')  # แสดงทศนิยม 4 ตำแหน่ง

    # การแยกผลลัพธ์เป็นชายหรือหญิง
    if prediction_value > 0.5:
        return 'Male'  # หากค่า > 0.5 ทำนายเป็นผู้ชาย
    else:
        return 'Female'  # หากค่า <= 0.5 ทำนายเป็นผู้หญิง

img_path = 'test4.jpg'  # ใส่ที่อยู่ไฟล์ภาพที่ต้องการทำนาย
predicted_gender = predict_gender(img_path)
print(f'Predicted Gender: {predicted_gender}')
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.applications.mobilenet import MobileNet, preprocess_input
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.models import Model

# Load the MobileNet model pre-trained on ImageNet, excluding the top layer
base_model = MobileNet(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Add custom layers for gender classification
x = Flatten()(base_model.output)
x = Dense(512, activation='relu')(x)
x = Dense(1, activation='sigmoid')(x)  # Output: 1 for binary classification (Male/Female)

# Final model
model = Model(inputs=base_model.input, outputs=x)

# Freeze the base layers
for layer in base_model.layers:
    layer.trainable = False

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Function to prepare the image for prediction
def prepare_image(img_path):
    img = image.load_img(img_path, target_size=(224, 224))  # Resize the image to 224x224
    img_array = image.img_to_array(img)  # Convert image to array
    img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension
    img_array = preprocess_input(img_array)  # Preprocess the image for MobileNet
    return img_array

# Function to predict gender
def predict_gender(img_path):
    img_array = prepare_image(img_path)

    # Predict the gender
    prediction = model.predict(img_array)
    prediction_value = prediction[0][0]  # Get the predicted value

    # Show the image
    img = image.load_img(img_path)
    plt.imshow(img)
    plt.axis('off')  # Hide axes
    plt.show()

    # Print the prediction value
    print(f'Prediction Value: {prediction_value:.4f}')  # Display prediction value

    # Return gender based on prediction value
    if prediction_value > 0.5:
        return 'Male'  # If prediction value is greater than 0.5, classify as Male
    else:
        return 'Female'  # If prediction value is less than or equal to 0.5, classify as Female

# Example usage
img_path = 'test4.jpg'  # Replace with your image path
predicted_gender = predict_gender(img_path)
print(f'Predicted Gender: {predicted_gender}')
```
